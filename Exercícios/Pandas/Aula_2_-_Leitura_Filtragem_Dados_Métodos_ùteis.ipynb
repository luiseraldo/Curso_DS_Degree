Conhecendo as bases responda:

a) Quais foram os produtos vendidos? 


```python
customers = pd.read_csv(".\data\customer.csv")
products = pd.read_csv(".\data\products.csv")
sales = pd.read_csv(".\data\sales.csv")



sales.set_index("sale_id", inplace=True)
products.set_index("p_id", inplace=True)
customers.set_index("c_id", inplace=True)



p_id_vendidos = sales["p_id"].unique()

produtos_vendidos = products.loc[p_id_vendidos]

produtos_vendidos
```

b) E os não vendidos?

```python
#b)

Prod_n_vendidos = products.drop(p_id_vendidos)

Prod_n_vendidos

```

c) Quais foram as cinco maiores vendas? Salve essas vendas em um arquivo excel.

```python




for sale in sales.index:
    sales.loc[sale, "valor_total"] = products.loc[sales.loc[sale, "p_id"],"price"] * sales.loc[sale, "qty"]
    
sales.sort_values("valor_total", inplace=True, ascending=False)

sales.head(5)
```
